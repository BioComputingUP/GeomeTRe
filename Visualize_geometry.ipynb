```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
results_s=pd.read_csv('Outputs.3/Nice_solenoids.csv')
results_p=pd.read_csv('Outputs.3/Nice_propellers.csv')
results_s2=pd.read_csv('Outputs.3/Irregular_solenoids.csv')
```

```python
mask1=results_s['unit start'] != 'mean'
mask2=results_s['unit start'] != 'std deviation'
mask3=results_s['Unnamed: 0'] != 0
data_s=results_s[mask1 & mask2 & mask3]

mask1=results_p['unit start'] != 'mean'
mask2=results_p['unit start'] != 'std deviation'
mask3=results_p['Unnamed: 0'] != 0
data_p=results_p[mask1 & mask2 & mask3]
mask1=results_s2['unit start'] != 'mean'
mask2=results_s2['unit start'] != 'std deviation'
mask3=results_s2['Unnamed: 0'] != 0

data_s2=results_s2[mask1 & mask2 & mask3]
```

```python
means_s=results_s[results_s['unit start'] == 'mean']
deviations_s=results_s[results_s['unit start'] == 'std deviation']
means_p=results_p[results_p['unit start'] == 'mean']
deviations_p=results_p[results_p['unit start'] == 'std deviation']
means_s2=results_s2[results_s2['unit start'] == 'mean']
deviations_s2=results_s2[results_s2['unit start'] == 'std deviation']
```

PROTEIN LEVEL RESULTS

Twist

```python
fig,ax=plt.subplots(1,2)
ax[0].hist(means_s['twist'],density=True,label='Solenoids')
ax[0].hist(means_p['twist'],density=True,alpha=0.7,label='Propellers')
ax[0].hist(means_s2['twist'],density=True,alpha=0.7,label='Irregular Solenoids')
ax[0].set_title('twist mean')
ax[0].set_ylabel('Frequency')
ax[0].legend()
ax[1].hist(deviations_s['twist'],density=True,label='Solenoids')
ax[1].hist(deviations_p['twist'],density=True,alpha=0.7,label='Propellers')
ax[1].hist(deviations_s2['twist'],density=True,alpha=0.7,label='Irregular Solenoids')
ax[1].set_title('twist std')
ax[1].set_ylabel('Frequency')
ax[1].legend()
```

```python
plt.scatter(means_s['twist'],deviations_s['twist'],label='Solenoids')
plt.scatter(means_p['twist'],deviations_p['twist'],label='Propellers')
plt.scatter(means_s2['twist'],deviations_s2['twist'],label=' Irregular Solenoids')
plt.xlabel('Mean')
plt.ylabel('Std')
plt.title('Twist')
plt.legend()
```
